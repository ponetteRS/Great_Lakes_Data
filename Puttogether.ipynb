```python
import requests as re
import json
import os
import sqlite3
import pandas as pd
```

```python
#read glansis tracking into csv, has species list to query
tracking = pd.read_csv('Tracking.csv')
#lower case titles
tracking_clean =tracking.rename(str.lower, axis='columns', inplace = False)
#save 'scientific name' column in tracking as a list called species to be base of each query
species=tracking_clean['scientific name'].to_list()
```

```python
#read GLANSIS bulk upload template, has necessary columns to search for
output_table = pd.read_csv('bulk.csv')
```

```python
#narrows output table df to only mandatory columns
output_table=output_table[['Genus*', 'Species*', 'Latitude*', 'Source*', 'Accuracy*', 'Locality*', 'scientific_name']]
```

```python
def make_a_query(query, key):
    base_url = 'https://api.si.edu/openaccess/api/v1.0/search'
    params_dict= {'api_key':key, 'q':query}
    first_search = re.get(base_url, params=params_dict)
    first_result =json.loads(first_search.text)
    return first_result 
#searches for a species with the Smithsonian 'search' API
```

```python
#to get the scientific name of a speices
def get_scientific_name(first_result):
    list_of_titles = []
    for i in range(len(first_result['response']['rows'])):
        title= first_result['response']['rows'][i]['content']['descriptiveNonRepeating']['title']['content']
        title_string=json.dumps(title)
        list_of_titles.append(title_string) #add each name to scientific_name in df
    titles_series= pd.Series(list_of_titles)
    output_table['scientific_name'] = titles_series
    #for i in titles_series:
        #output_table['scientific_name'].append(i)
    #print(list_of_titles)
        #output_table.at(1, 'scientific_name') 
    return None
```

```python
def main():
    for i in species:
        cool = make_a_query(i, '11icycZDd63MfM3OrfwyQwrpJatRz5mFy37cnBo6') #saved as a json object. it is the file
        get_scientific_name(cool)
    output_table
        
        
#mainn called
if __name__ == "__main__":
    main()        
```
