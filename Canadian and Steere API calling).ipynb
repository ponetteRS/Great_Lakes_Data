```
# searches for a species with the 'search' API returns the text of the whole query result
# def make_a_query(key,query):
def make_a_query(query):
    base_url = 'https://api.gbif.org/v1/species/search'
    # params_dict={'api_key':key, 'q':query}
    params_dict={'q':query}
    first_search = re.get(base_url, params_dict)
    first_result = json.loads(first_search.text)
    return first_result

# testing function
#make_a_query(["Cabomba caroliniana"])
```

```
# Genus, Species, Latitude, Longitud,Source, Accuracy, Locality, year,Record Type
make_a_query(["Cabomba caroliniana"])
```

```
def species(lst):
    url= 'https://api.gbif.org/v1/species'
    final = []
    for l in lst:
        # param = {"language": language, "datasetKey": datasetKey, "sourceId": sourceId, "name": name}  
        param = {"name": l}  
        response = requests.get(url, params= param)
        response_data = response.json()
        final.append(response_data)
    return final
```

```
lst = ["Cabomba caroliniana"]
species(lst)
```
